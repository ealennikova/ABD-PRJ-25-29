```python
# Словарь признаков (с синонимами)
features_rus = {
    "частота дыхания": ["дыхание", "дышать", "дыхательный"],
    "температура тела": ["температура", "градусы", "тепло"],
    "движение конечностей": ["движение", "активность", "моторика"],
    "сатурация": ["сатурация", "кислород", "SpO2"],
    "движение глаз": ["глаз", "зрение", "REM"],
    "часы сна": ["сон", "спать", "отдых"],
    "пульс": ["пульс", "сердце", "сердечный"],
    "уровень стресса": ["стресс", "напряжение", "возбуждение"]
}

# Проверяем есть ли текст
if text_block:
    # Очищаем текст
    raw_text = text_block.get_text(separator=" ").strip()
    cleaned_text = re.sub(r'\s+', ' ', raw_text)  # Убираем лишние пробелы

    # Разбиваем на абзацы по "<hr>"
    paragraphs = cleaned_text.split("<hr>")

    found_features = {feature: [] for feature in features_rus}

    # Проходим по каждому абзацу
    for paragraph in paragraphs:
        # Лемматизируем слова абзаца
        words = re.findall(r'\b\w+\b', paragraph.lower())
        words_lemmas = [morph.parse(word)[0].normal_form for word in words]

        # Ищем совпадения
        for feature, synonyms in features_rus.items():
            if any(morph.parse(syn)[0].normal_form in words_lemmas for syn in synonyms):
                found_features[feature].append(paragraph.strip())

    # Сохраняем в CSV
    for feature, contexts in found_features.items():
        if contexts:  # Только если есть совпадения
            df = pd.DataFrame({"Абзац": contexts})
            filename = f"{feature}.csv"
            df.to_csv(filename, index=False, encoding="utf-8-sig")  # Исправлена кодировка
            print(f"Найдено {len(contexts)} совпадений для '{feature}', сохранено в {filename}")
        else:
            print(f"Ничего не найдено для '{feature}'")

else:
    print("Не удалось найти текст книги.")
```
