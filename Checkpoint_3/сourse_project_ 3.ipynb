```python
В итоге что делаем:
body temperature	Обрезать выбросы по IQR и заменить медианой
hours of sleep	    Аналогично- обрезать по IQR
blood oxygen	    Все, что выше 100%, приравнять к 100% (медицинская логика)
Остальные	        Пока оставить как есть, но можно зафиксировать в выводах, что есть крайние случаи
```

```python
def clip_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
    return df


columns_with_outliers = ['body temperature', 'hours of sleep']
for col in columns_with_outliers:
    df = clip_outliers_iqr(df, col)
```

```python
from scipy.stats import ttest_ind, levene
```

```python
df_main = pd.read_csv(r"C:\course_project\datasets\data_main.csv")
df_vlad = pd.read_csv(r"C:\course_project\datasets\data_Vlad.csv")
df_elena = pd.read_csv(r"C:\course_project\datasets\data_Elena.csv")
df_alex = pd.read_csv(r"C:\course_project\datasets\data_Alex.csv")

# Приведение всех названий колонок к единому виду
for df in [df_main, df_vlad, df_elena, df_alex]:
    df.columns = df.columns.str.strip().str.lower()
```

```python
# Список признаков для анализа кроме stress levels, так как он категориальный
features = [
    'snoring range', 'respiration rate', 'body temperature',
    'limb movement', 'blood oxygen', 'eye movement',
    'hours of sleep', 'heart rate']
```

```python
results = []

datasets = {
    "vlad": df_vlad,
    "elena": df_elena,
    "alex": df_alex}

for name, df in datasets.items():
    for feature in features:
        
        # t-test
        t_stat, p_ttest = ttest_ind(df_main[feature], df[feature], nan_policy='omit')

        # Levene's test (на гомогенность дисперсий)
        l_stat, p_levene = levene(df_main[feature], df[feature], center='mean')

        results.append({
            "сравнение": f"main vs {name}",
            "признак": feature,
            "p-value t-test": round(p_ttest, 4),
            "p-value levene": round(p_levene, 4)
        })

# Преобразуем в таблицу
results_df = pd.DataFrame(results)

display(results_df[results_df['сравнение'] == 'main vs vlad'])
```
